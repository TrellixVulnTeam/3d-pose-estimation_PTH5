```python
%matplotlib inline

import os
import h5py
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.lines import Line2D

path = '/home/fpena/datasets/h36m/processed/S1/Directions-1/'
annot = h5py.File(path + 'annot.h5', 'r')
```

```python
cameras = ['54138969', '55011271', '58860488', '60457274']
camera_idx = 1
frame_idx = 300
idx = 0
while str(annot['camera'][idx]) != cameras[camera_idx]:
    idx += 1
    pass
idx += frame_idx
```

```python
annot['camera'][0]
```

```python
pjoin = os.path.join

image_filename = 'img_{:06d}.jpg'.format(annot['frame'][idx])
img = mpimg.imread(pjoin(path, 'imageSequence', cameras[camera_idx], image_filename))
markers_position = annot['pose']['2d'][idx]
markers = [0, 1, 2, 3, 4, 6, 7, 8, 9, 12, 13, 14, 17, 18, 19, 25, 26, 27]
segments = [
    [0, 1],
    [0, 6],
    [1, 2],
    [2, 3],
    [3, 4],
    [6, 7],
    [7, 8],
    [8, 9],
    [0, 12],
    [12, 13],
    [13, 14],
    [13, 17],
    [17, 18],
    [18, 19],
    [13, 25],
    [25, 26],
    [26, 27],
]

(len(markers),)
```

```python
fig, ax = plt.subplots()
fig.set_figheight(30)
fig.set_figwidth(20)

# Show image
ax.imshow(img)

# Show markers
ax.scatter(markers_position[markers,0], markers_position[markers,1])

# Show skeleton
for segment in segments:
    x1, y1 = markers_position[segment[0]]
    x2, y2 = markers_position[segment[1]]
    l = Line2D([x1, x2], [y1, y2], color = "red", linestyle = "dashed")
    ax.add_line(l)
    pass

# Show marker ids
for i in range(32):
    if not i in markers:
        continue
    ax.text(markers_position[i, 0], markers_position[i, 1], str(i), size=16, color='green')
    pass

plt.show()
```
